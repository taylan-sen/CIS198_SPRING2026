### github notes

key terms:
* **repository** - a container holding all *committed* files and folders.
* **commit** - saving a file version in a repository (along with a save message, i.e. commit message)
* **git** - version control program (i.e. a app for saving many versions of files, ex. computer code.
* **github.com** - a online company (website) providing a place for users to save their repositories.

### google search notes

**double quotes** - when you put words or phrases in double quotes, this *forces* google to return only pages that have the actual *string* in double quoutes

**site:github.com** - using the *site:* keyword limits the search results to the domain name (URL) provided.

example: to search for pages about tourism on niagara's website you use the search phrase:

    site:niagara.edu tourism




### Python notes
