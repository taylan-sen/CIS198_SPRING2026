# 🧠 Introduction to Python Sets

Sets are an essential Python data structure used to store **unique elements** and perform **fast membership tests**.

In this notebook, you'll learn:
- What sets are and why they are useful
- How to create and use them
- Common operations
- Practice questions


## 🚀 Motivation

Imagine you have a list with duplicate values:
```python
[1,2,2,3,3,3]
```
What if you only care about **unique values**?

Or you want to quickly check:
- Is a value in a list?
- What items are shared between two lists?

👉 Sets are designed for exactly these tasks.


## 🔧 Creating Sets


````python
a = {1,2,3}
print(a)

# From a list (removes duplicates)
b = set([1,2,2,3,3])
print(b)
````

## ⚡ Fast Membership Testing


````python
my_set = {1,2,3,4,5}

print(3 in my_set)  # True
print(10 in my_set) # False
````

## 🔄 Common Set Operations

- Union (combine)
- Intersection (common elements)
- Difference (what's unique)


````python
a = {1,2,3}
b = {3,4,5}

print(a | b)  # Union
print(a & b)  # Intersection
print(a - b)  # Difference
````

## 🧪 Example: Removing Duplicates


````python
numbers = [1,2,2,3,4,4,5]
unique_numbers = set(numbers)
print(unique_numbers)
````

## ➕ Adding Elements to Sets

Python gives you a few ways to add elements to a set:

### `add()` — Add a single element
Use `.add()` to insert one item at a time. If the element already exists, the set stays unchanged (no duplicates!).

```python
my_set = {1, 2, 3}
my_set.add(4)    # {1, 2, 3, 4}
my_set.add(2)    # {1, 2, 3, 4} — no change, 2 already exists
```

### `update()` — Add multiple elements at once
Use `.update()` to merge in elements from another iterable (list, tuple, string, or another set).

```python
my_set = {1, 2}
my_set.update([3, 4, 5])        # {1, 2, 3, 4, 5}
my_set.update({6}, (7,), "ab")  # also adds 6, 7, 'a', 'b'
```

### Union operator `|=` — Combine sets in place
The `|=` operator works like `.update()` but only accepts other sets.

```python
a = {1, 2}
b = {3, 4}
a |= b   # a is now {1, 2, 3, 4}
```

### Key things to remember
- You **cannot** add mutable objects (like lists or dictionaries) to a set — elements must be **hashable**.
- Adding a duplicate element is silently ignored; it never raises an error.
- `add()` is for **one element**, `update()` is for **many elements**.

````python
# --- add(): insert a single element ---
colors = {"red", "green", "blue"}
colors.add("yellow")
print("After add('yellow'):", colors)

colors.add("red")
print("After add('red') again:", colors)  # no duplicate added

# --- update(): insert multiple elements from an iterable ---
colors.update(["orange", "purple"])
print("After update(['orange', 'purple']):", colors)

colors.update({"pink"}, ("white",))
print("After update with set and tuple:", colors)

# --- |= operator: merge another set in place ---
primary = {"red", "blue", "yellow"}
secondary = {"green", "orange", "purple"}
primary |= secondary
print("After |= merge:", primary)

# --- Attempting to add an unhashable type raises an error ---
try:
    colors.add([1, 2, 3])
except TypeError as e:
    print(f"Cannot add a list: {e}")
````
